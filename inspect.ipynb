```python
from pathlib import Path

from ocr.read import PyMuPDFPageOCR
from layout.order import heuristic_reading_order
from layout.detect import PaddleLayoutDetector

from utils.visualize import visualize_ocr, visualize_layout_detection
```

```python
MIN_CONFIDENCE = 0.5
PAGE_INDEX = 9  

PDF_PATH = Path("data/raw/dnd_rulebook.pdf")
IMAGE_PATH = Path(f"data/intermediate/pages/page_{PAGE_INDEX:04d}.png")
RENDER_DIR = Path("data/intermediate/pages/")
```

```python
ocr = PyMuPDFPageOCR()
layout_detector = PaddleLayoutDetector(min_confidence=MIN_CONFIDENCE)
```

```python
ocr_regions = ocr.ocr_page(PDF_PATH, PAGE_INDEX-1)
reading_order = heuristic_reading_order(ocr_regions, 2)
layout_regions = layout_detector.detect_page(IMAGE_PATH)
```

```python
visualize_ocr(IMAGE_PATH, ocr_regions, reading_order)
```

```python
visualize_layout_detection(IMAGE_PATH, layout_regions)
```
